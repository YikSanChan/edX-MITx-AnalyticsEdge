以下代码必须同时在同一个cell中执行，因为每个cell都会生成一个新的plot视图。

```r
plot(CocaCola$Date, CocaCola$StockPrice, type="l",col="red")#Problem 2.1
lines(ProcterGamble$Date, ProcterGamble$StockPrice, col="blue")
abline(v=as.Date(c("2000-03-01")), lwd=2)
abline(v=as.Date(c("1983-01-01")), lwd=2)
```

```r
plot(CocaCola$Date[301:432], CocaCola$StockPrice[301:432], type="l", col="red", ylim=c(0,210))
lines(ProcterGamble$Date[301:432], ProcterGamble$StockPrice[301:432], col="blue")
lines(IBM$Date[301:432], IBM$StockPrice[301:432], col="green")
lines(GE$Date[301:432], GE$StockPrice[301:432], col="purple")
lines(Boeing$Date[301:432], Boeing$StockPrice[301:432], col="orange")
abline(v=as.Date(c("2000-03-01")), lwd=2)
abline(v=as.Date(c("1997-09-01")), lwd=2, lty=3)
abline(v=as.Date(c("1997-11-01")), lwd=2, lty=3)
abline(v=as.Date(c("2004-01-01")), lwd=2, lty=3)
abline(v=as.Date(c("2005-12-01")), lwd=2, lty=3)
```
